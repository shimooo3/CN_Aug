```python
import os
import json
import random

source_path = '../img_aug/source/'
target_path = '../img_aug/target/'

source_files = os.listdir(source_path)
target_files = os.listdir(target_path)

prompts = [
    "prompt A",
    "prompt B",
    "prompt C",
    "prompt D",
    "prompt E",
]


data = []

for file_name in source_files:
    if file_name in target_files:
        entry = {
            "source": f"source/{file_name}",
            "target": f"target/{file_name}",
            # choose randomly from prompts
            "prompt": random.choice(prompts)
        }
        data.append(entry)

json_string = json.dumps(data, indent=4)
json_string_without_brackets = json_string[1:-1].strip()

# 各エントリの"}"の後の","を削除する
lines = json_string_without_brackets.split('\n')
cleaned_lines = []
for line in lines:
    if line.strip() == '},':
        cleaned_lines.append('}')
    else:
        cleaned_lines.append(line)

cleaned_json_string = '\n'.join(cleaned_lines)


# JSONファイルに書き込む
with open('prompt.json', 'w') as json_file:
    json_file.write(cleaned_json_string)

print("JSONファイルが作成されました。")
```
